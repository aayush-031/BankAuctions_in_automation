```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
```

```python
def dataval(tables):
    fin_dfs =pd.DataFrame()
    for i in tables:
        temp_df =pd.DataFrame()
        try:
            temp_df=pd.read_html(str(i))[0]
        except:
            pass
        if temp_df.shape[1]>2:
            fin_dfs=temp_df.copy()
    return fin_dfs
```

```python
def scrapweb(url):
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source,'lxml')
    tables = soup.find_all('table')
    fin_dfs=pd.DataFrame()
    fin_dfs=dataval(tables)
    all_links = soup.find_all('a', href=True)
    view_links = []
    for link in all_links:
        href = link['href']
        if '/auction/' in href:
            view_links.append(href)
    print(view_links)
    print(len(view_links))
    fin_dfs['VIEW']=view_links
    input_tag = driver.find_element_by_id('home-page-table_next')
    nextButtonClass =(input_tag.get_attribute('class'))
    while("disabled" not in nextButtonClass):
        driver.find_element_by_id('home-page-table_next').click()
#         time.sleep(5)
        soup = BeautifulSoup(driver.page_source,'lxml')
        tables = soup.find_all('table')
        fin_df =pd.DataFrame()
        fin_df=dataval(tables)
        all_links = soup.find_all('a', href=True)
        view_links = []
        for link in all_links:
            href = link['href']
            if '/auction/' in href:
                view_links.append(href)
        print(view_links)
        print(len(view_links))
        fin_df['VIEW']=view_links
        fin_dfs=fin_dfs.append(fin_df)
        nextButtonClass=driver.find_element_by_id('home-page-table_next').get_attribute('class')
    driver.close()
    return fin_dfs
```

```python
dfs = scrapweb('https://bankauctions.in/')
```

```python
dfs['Reserve Price']=dfs['Reserve Price'].replace('₹','',regex=True)
```

```python
dfs['Reserve Price']=dfs['Reserve Price'].replace(',','',regex=True)
dfs.drop('View',axis=1,inplace=True)
```
